```python
import numpy as np
from scipy.optimize import minimize


def f(params):
    x, y = params
    return (
        (1.5 - x + x * y) ** 2
        + (2.25 - x + x * y**2) ** 2
        + (2.625 - x + x * y**3) ** 2
    )


initial_guess = [0, 0]
bounds = [(-4.5, 4.5), (-4.5, 4.5)]

result = minimize(f, initial_guess, bounds=bounds)
min_x, min_y = result.x
min_value = result.fun

print(f"Minimum value: {min_value}")
print(f"Minimum point: x = {min_x}, y = {min_y}")
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Load data from CSV
data_q2 = np.genfromtxt("data_q2.csv", delimiter=",")
```

```python
plt.hist(data_q2, bins=20, density=True, alpha=0.6, label="Histogram of Data")

mu, sigma = np.mean(data_q2), np.std(data_q2)
x = np.linspace(min(data_q2), max(data_q2), 100)
gaussian = norm.pdf(x, mu, sigma)
plt.plot(x, gaussian, color="red", label="Gaussian Distribution")

plt.legend()
plt.xlabel("Value")
plt.ylabel("Density")
plt.title("Histogram and Gaussian Distribution")
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# Provided data
x_values = np.array(
    [
        -1,
        -0.777777778,
        -0.555555556,
        -0.333333333,
        -0.111111111,
        0.111111111,
        0.333333333,
        0.555555556,
        0.777777778,
        1,
    ]
)

y_values = np.array(
    [
        -0.52398042,
        -0.431469165,
        -0.354161175,
        -0.195195461,
        -0.158500249,
        -0.020306262,
        0.030828491,
        0.11471506,
        0.279141306,
        0.449739825,
    ]
)

# Calculate means
x_mean = np.mean(x_values)
y_mean = np.mean(y_values)

# Calculate beta1
beta1 = np.sum((x_values - x_mean) * (y_values - y_mean)) / np.sum(
    (x_values - x_mean) ** 2
)

# Calculate beta0
beta0 = y_mean - beta1 * x_mean

print(f"Estimated beta0: {beta0}")
print(f"Estimated beta1: {beta1}")

# Regression line
regression_line = beta0 + beta1 * x_values

# Plotting
plt.scatter(x_values, y_values, label="Data")
plt.plot(x_values, regression_line, color="red", label="Regression Line")
plt.xlabel("x")
plt.ylabel("y")
plt.title("Data and Regression Line")
plt.legend()
plt.show()
```
